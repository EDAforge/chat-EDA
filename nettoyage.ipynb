##### Breed - Correction

```python
data['Breed'].value_counts()
```

```python
data.loc[data['Breed'].isin(['wrack doll','rack doll','ragdoll']), 'Breed'] =  'Ragdoll'
data.loc[data['Breed'].isin(['My coon','Maine loon','maine coon']),'Breed'] = 'Maine coon'
data.loc[data['Breed'].isin(['Ankora','Angorra']), 'Breed'] = 'Angora'

```

#### Age - correction

```python
column = ['Age_in_months', 'Age_in_years']
filtre = (data['Age_in_months'] < 0) | (data['Age_in_years'] < 0)
data.loc[filtre, column] = data.loc[filtre, column].abs()
data
```

```python
data[['Age_in_months', 'Age_in_years']].dtypes
```

```python
data[['Age_in_months', 'Age_in_years']] = data[['Age_in_months', 'Age_in_years']].astype('int16')
```
